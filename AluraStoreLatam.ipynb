### Importación de datos



```python
import pandas as pd

url1 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda1 = pd.read_csv(url1)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda1.head(10)
```

#1. Análisis de facturación



```python
# Calculo del ingreso total de cada tienda sumando la columna 'Precio' de cada tienda
ingreso_tienda1 = tienda1['Precio'].sum()
ingreso_tienda2 = tienda2['Precio'].sum()
ingreso_tienda3 = tienda3['Precio'].sum()
ingreso_tienda4 = tienda4['Precio'].sum()

# Se muestran los resultados
print(f"Ingreso total de Tienda 1: ${ingreso_tienda1:,.2f}")
print(f"Ingreso total de Tienda 2: ${ingreso_tienda2:,.2f}")
print(f"Ingreso total de Tienda 3: ${ingreso_tienda3:,.2f}")
print(f"Ingreso total de Tienda 4: ${ingreso_tienda4:,.2f}")
```

# 2. Ventas por categoría

```python
# Calculo de la cantidad de productos vendidos por categoría en cada tienda
ventas_cat_tienda1 = tienda1.groupby('Categoría del Producto').size().reset_index(name='Cantidad')
ventas_cat_tienda2 = tienda2.groupby('Categoría del Producto').size().reset_index(name='Cantidad')
ventas_cat_tienda3 = tienda3.groupby('Categoría del Producto').size().reset_index(name='Cantidad')
ventas_cat_tienda4 = tienda4.groupby('Categoría del Producto').size().reset_index(name='Cantidad')

# Mostrar los resultados para cada tienda
print("\nTienda 1 - Ventas No. articulos por categoría:")
print(ventas_cat_tienda1.to_string(index=False))
print("\nTienda 2 - Ventas No. articulos por categoría:")
print(ventas_cat_tienda2.to_string(index=False))
print("\nTienda 3 - Ventas No. articulos por categoría:")
print(ventas_cat_tienda3.to_string(index=False))
print("\nTienda 4 - Ventas No. articulos por categoría:")
print(ventas_cat_tienda4.to_string(index=False))
```

## Ingresos por Categoría

```python
# Calculo de los ingresos por categoría en cada tienda
ingreso_cat_tienda1 = tienda1.groupby('Categoría del Producto')['Precio'].sum().reset_index(name='Total Ventas')
ingreso_cat_tienda2 = tienda2.groupby('Categoría del Producto')['Precio'].sum().reset_index(name='Total Ventas')
ingreso_cat_tienda3 = tienda3.groupby('Categoría del Producto')['Precio'].sum().reset_index(name='Total Ventas')
ingreso_cat_tienda4 = tienda4.groupby('Categoría del Producto')['Precio'].sum().reset_index(name='Total Ventas')

# Mostrar los resultados para cada tienda con formato
print("\nTienda 1 - Ingresos por categoría:")
print(ingreso_cat_tienda1.to_string(index=False, formatters={'Total Ventas': '${:,.2f}'.format}))
print("\nTienda 2 - Ingresos por categoría:")
print(ingreso_cat_tienda2.to_string(index=False, formatters={'Total Ventas': '${:,.2f}'.format}))
print("\nTienda 3 - Ingresos por categoría:")
print(ingreso_cat_tienda3.to_string(index=False, formatters={'Total Ventas': '${:,.2f}'.format}))
print("\nTienda 4 - Ingresos por categoría:")
print(ingreso_cat_tienda4.to_string(index=False, formatters={'Total Ventas': '${:,.2f}'.format}))
```

# 3. Calificación promedio de la tienda

# 4. Productos más y menos vendidos

## Los 5 productos más y menos vendidos por tienda

```python
# Lista de tiendas y nombres
tiendas = [(tienda1, 'Tienda 1'), (tienda2, 'Tienda 2'), (tienda3, 'Tienda 3'), (tienda4, 'Tienda 4')]

# Diccionarios para almacenar los resultados por tienda
ventas_prod_por_tienda = {}
top5_por_tienda = {}
bottom5_por_tienda = {}


for tienda_df, tienda_nombre in tiendas:
  ventas_prod = tienda_df.groupby('Producto').size().reset_index(name='Ventas')
  top5 = ventas_prod.sort_values('Ventas', ascending=False).head(5)
  bottom5 = ventas_prod.sort_values('Ventas', ascending=True).head(5)

 # Almacenar los resultados en los diccionarios
  ventas_prod_por_tienda[tienda_nombre] = ventas_prod
  top5_por_tienda[tienda_nombre] = top5
  bottom5_por_tienda[tienda_nombre] = bottom5

  print(f"\n{tienda_nombre} - Top 5 productos más vendidos:")
  print(top5.to_string(index=False))
  print(f"\n{tienda_nombre} - 5 productos menos vendidos:")
  print(bottom5.to_string(index=False))
```

## Comparativa de ingresos totales por tienda

```python
import numpy as np
# Crear listas para los datos
tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ingresos = [ingreso_tienda1, ingreso_tienda2, ingreso_tienda3, ingreso_tienda4]

# Convertir la lista de ingresos a un array de NumPy para realizar la división elemento a elemento
ingresos_millones = np.array(ingresos) / 1_000_000

# Crear gráfico de barras
plt.figure(figsize=(8, 6))
bars = plt.bar(tiendas_nombres, ingresos_millones, color='#1f77b4')

# Configurar título y etiquetas
plt.title('Ingresos Totales por Tienda')
plt.xlabel('Tienda')
plt.ylabel('Ingresos (Millones $)')

# Formatear eje Y en millones con 2 decimales
plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:.2f}'))

# Quitar marco superior y derecho
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Agregar cuadrícula en el eje Y
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Mostrar valores encima de las barras (en millones, 2 decimales)
for i, ingreso in enumerate(ingresos_millones):
    plt.text(i, ingreso + 0.02 * ingreso, f'{ingreso:.2f}', ha='center', va='bottom')

# Mostrar gráfico
plt.show()
```

## Participación Porcentual de Ingresos por Tienda en el total general

```python
# Calcular el total general de ingresos
total_ingresos = sum(ingresos)

# Calcular porcentajes
porcentajes = [(ingreso / total_ingresos) * 100 for ingreso in ingresos]

# Crear gráfico de torta
plt.figure(figsize=(8, 8))
plt.pie(porcentajes, labels=tiendas_nombres, autopct='%1.1f%%', colors=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'], startangle=90)
plt.title('Participación Porcentual de Ingresos por Tienda en el total general')

# Asegurar que el gráfico sea un círculo
plt.axis('equal')

# Mostrar gráfico
plt.show()
```
